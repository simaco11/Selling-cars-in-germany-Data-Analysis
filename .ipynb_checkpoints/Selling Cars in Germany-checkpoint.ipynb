## About

This dataset is downloaded from [kaggle.com]('https://www.kaggle.com/ander289386/cars-germany').
The data is collected from AutoScout24, one of the largest Europe's car market for new and used cars. 
\
The dataset contains a list of cars that are selling in Germany for the past 10 years.

### Importing Libraries and Loading the Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Setting styles for charts

```python
from matplotlib import rcParams

sns.set_style('darkgrid')
rcParams['figure.figsize'] = 10.7,7.27
```

### Reading The Data

```python
cars_df = pd.read_csv('cars-germany-dataset.csv')
```

```python
cars_df.shape
```

```python
cars_df.describe()
```

### Data Cleaning

```python
cars_df.head()
```

```python
cars_df.isna().sum()
```

### Dealing with NaN values

```python
cars_df = cars_df.dropna()
```

```python
cars_df.isna().sum()
```

Now that we are cleaned our data from NaN values we are ready to start with data analysis.

## Analysing and Visualising data  

### Q1: Which car manufacturers sold the most cars in Germany in the past 10 years?

```python
cars_df['make'].value_counts().head(10)
```

#### Let's visualise this by Seaborn's Countplot

```python
sns.countplot(x= 'make', data = cars_df, order=cars_df.make.value_counts().iloc[:10].index);
plt.xticks(rotation=50)
plt.ylabel('Number of sold cars')
plt.xlabel('Manufacturers');
```

### Q2: Which gearbox type is most common in selling cars?

```python
gearbox = cars_df['gear'].value_counts();
gearbox.head()
```

```python
explode = (0.1, 0, 0)
plt.pie(gearbox, labels=gearbox.index, explode = explode,shadow = True, autopct='%1.1f%%', startangle=180)
plt.title('Gearbox frequency');
```

### Q3: What Fuel is most common in selling cars?

```python
fuel_type_pct = cars_df['fuel'].value_counts() / cars_df['fuel'].count() *100
fuel_type_pct = fuel_type_pct.head(5).round(2)
fuel_type_pct
```

```python
explode = (0.1, 0, 0, 0, 0)
plt.pie(fuel_type_pct, labels = fuel_type_pct.index, explode = explode, shadow = True, autopct='%1.1f%%', startangle = 180);
plt.title('Fuel Type');
```

### Q3: Which are the 10 most expensive cars sold in the last 10 years?

```python
most_expensive = cars_df[['make','model', 'price']].groupby('price').sum().sort_values('price', ascending = False).head(10)
most_expensive['make_and_model'] = most_expensive['make'] + ' ' + most_expensive['model']
most_expensive
```

```python
sns.barplot(x = most_expensive['make_and_model'], y = most_expensive.index);
plt.xticks(rotation = 80);
plt.title('Most expensive cars sold in the last 10 years');
```

### Q4: Which year was the best for selling cars?

```python
cars_per_year = cars_df.groupby('year')['make'].count()
```

```python
sns.barplot(x = cars_per_year.index, y = cars_per_year)
plt.ylabel('Sold Per Year')
plt.title('Cars Sold Per Year');
```

### Q5: How much price of the cars was increased in the last 10 years?

```python
price_per_year = cars_df.groupby('year').sum()
price_per_year['price'].head(11)
```

```python
plt.plot(price_per_year.index, price_per_year['price'], marker = 'o',mew = 5, mec = 'red');
plt.xticks(price_per_year.index);
plt.xlabel('Year')
plt.ylabel('Price in USD ($)')
plt.title('Total price of the cars per year');
```

### Q6: What is the correlation between HP and Price?

```python
plt.figure(figsize=(16,10))
sns.scatterplot(x ='hp', y ='price', data = cars_df, hue = 'fuel', s=60, palette = 'tab10');
```

### Q7: What is the correlation between Fuel type and Price?

```python
sns.barplot(x = 'fuel', y = 'price', data = cars_df );
plt.xticks(rotation = 90);
```

### Q8: Hp per Manufacturer

```python
plt.figure(figsize=(16,10))
plot = sns.scatterplot(x = 'make', y = 'hp',hue = 'make', data = cars_df);
plot.legend_.remove()
plt.xticks(rotation = 90);
```

### Q9: What is the biggest Mileage per manufacturer?

```python
mil_per_manufacturer = cars_df.groupby('make')['mileage'].sum().sort_values(ascending = False).head(10)
mil_per_manufacturer
```

```python
sns.barplot(x = mil_per_manufacturer.index, y = mil_per_manufacturer);
```
